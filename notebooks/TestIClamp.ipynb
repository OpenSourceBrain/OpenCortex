```python
import opencortex.build as oc

nml_doc, network = oc.generate_network("TestIClamp")

oc.add_cell_and_channels(nml_doc, 'acnet2/pyr_4_sym.cell.nml', 'pyr_4_sym')
```

```python

popHH = oc.add_single_cell_population(network,
                                     'popHH',
                                     'pyr_4_sym',
                                     z=100)

pgHH = oc.add_pulse_generator(nml_doc,
                       id="pgHH",
                       delay="100ms",
                       duration="300ms",
                       amplitude="0.7nA")

oc.add_inputs_to_population(network,
                            "Stim1",
                            popHH,
                            pgHH.id,
                            all_cells=True)

```

```python

nml_file_name = '%s.net.nml'%network.id
oc.save_network(nml_doc, nml_file_name, validate=True)

lems_file = oc.generate_lems_simulation(nml_doc, 
                            network, 
                            nml_file_name, 
                            duration =      500, 
                            dt =            0.005)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline  
from pyneuroml import pynml
results = pynml.run_lems_with_jneuroml_neuron(lems_file, nogui=True, load_saved_data=True, plot=True)

```
